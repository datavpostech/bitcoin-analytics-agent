```python
# Receber do job 4 a informação do outro agente
analise = dbutils.jobs.taskValues.get(
    taskKey='agente_analise',
    key='analise'
)
```

```python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
import os

load_dotenv()

# lê o prompt
with open('prompts/agente_html.txt', 'r') as f:
    prompt_template = f.read()

# substitui o placeholder
prompt = prompt_template.replace('{analise}', analise)

# chama o Gemini
llm = ChatGoogleGenerativeAI(
    model='gemini-2.5-flash-lite',
    google_api_key=os.getenv('GEMINI_API_KEY')
)

resposta = llm.invoke(prompt)
html = resposta.content

# passa o HTML para a próxima task
dbutils.jobs.taskValues.set(key='html', value=html)

print(html)
```
